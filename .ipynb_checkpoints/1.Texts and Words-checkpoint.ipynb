```python
# 词频统计
fdist1 = FreqDist(text1)
print(fdist1)
```

```python
fdist1.most_common(20)
```

```python
# 单词的累计频率图
fdist1.plot(50, cumulative=True)
```

```python
# z只出现一次的单词
fdist1.hapaxes()
```

```python
# 筛选单词 长度大于15
V = set(text1)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```

```python
# 长度超过七个字符的单词，发生七次以上
fdist5 = FreqDist(text5)
sorted(w for w in set(text5) if len(w) > 7 and fdist5[w] > 7)
```

```python
# 经常出现的二元组可以认为是搭配短语 collocations
text4.collocations()
```

```python
# 除了计数单词在文本中出现的次数 
# 还可以计数其他的东西
# 计数文本中字长的分布，其中每个数字是文本中相应单词的长度
fdist = FreqDist(len(w) for w in text1)
print(fdist)
```

```python
fdist.most_common()
```

```python
print(fdist.max())
print(fdist[3])
print(fdist.freq(3))
```

fdist = FreqDist(samples)	创建包含给定样本的频率分布
fdist[sample] += 1	增加此样本的计数
fdist['monstrous']	count of the number of times a given sample occurred
fdist.freq('monstrous')	frequency of a given sample
fdist.N()	总样本数
fdist.most_common(n)	the n most common samples and their frequencies
for sample in fdist:	迭代样本
fdist.max()	sample with the greatest count
fdist.tabulate()	tabulate the frequency distribution
fdist.plot()	频率分布图形图
fdist.plot(cumulative=True)	频率分布的累积图
fdist1 |= fdist2	update fdist1 with counts from fdist2
fdist1 < fdist2	test if samples in fdist1 occur less frequently than in fdist2测试fdist1中的样本是否比fdist2中的频率低

s.startswith(t)	test if s starts with t
s.endswith(t)	test if s ends with t
t in s	test if t is a substring of s
s.islower()	test if s contains cased characters and all are lowercase
s.isupper()	test if s contains cased characters and all are uppercase
s.isalpha()	test if s is non-empty and all characters in s are alphabetic
s.isalnum()	test if s is non-empty and all characters in s are alphanumeric
s.isdigit()	test if s is non-empty and all characters in s are digits
s.istitle()	test if s contains cased characters and is titlecased (i.e. all words in s have initial capitals)
